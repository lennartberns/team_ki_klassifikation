## Datenaufbereitung

### Entfernung Spalte Antwort

```python
data_df = rawdata_df.drop('Antwort',axis=1)
data_df
```

### Entfernung Null-Values

```python
data_df = data_df.dropna()
```

```python
data_df = data_df.drop_duplicates()
```

```python
data_df
```

```python
data_df = data_df[data_df['Kategorien'].str.contains('Null')==False]
data_df
```

### Entfernung Sonderzeichen in Kommentaren

```python
df_kommentar = data_df['Kommentar'].str.replace('\W', ' ', regex=True)
data_df.Kommentar = df_kommentar
#data_df.to_excel("C:/Users/loedu/Documents/test.xlsx")
data_df
```

```python
data_without_spaces = data_df.replace(' +', ' ', regex=True)
data_without_spaces.Kommentar = data_without_spaces.Kommentar.str.lstrip()
data_without_spaces
```

### Entfernung leere Kommentare

```python
cleandata_df = data_without_spaces.query("Kommentar != '' ")
cleandata_df

# cleandata_df.to_excel("C:/Users/flori/Documents/test4.xlsx")
```

```python
cleandata_df['Kommentar'] = cleandata_df['Kommentar'].str.lower()
```

```python
# Englische ÜBersetzung löschen
cleandata_df['Kommentar'] = cleandata_df['Kommentar'].str.split('translated by google').str[0]
```

```python
cleandata_df.Kategorien = cleandata_df.Kategorien.str.replace(' ', '')
```

### Entfernung Spalte Kunde

```python
cleandata_df = cleandata_df.drop('Kunde',axis=1)
cleandata_df
```

### Export bereinigter Datensatz

```python
cleandata_df.to_excel('data/clean/Google_Rezensionen.xlsx')
```
